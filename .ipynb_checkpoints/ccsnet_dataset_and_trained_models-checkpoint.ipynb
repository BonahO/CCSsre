```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning) 
import numpy as np
import os 
import h5py
import scipy.io
import matplotlib.pyplot as plt
import math
from tensorflow import keras
import tensorflow as tf
from ReflectionPadding3D import ReflectionPadding3D
```

```python
def get_dataset(name):
    hf_r = h5py.File(f'/data/cees/gegewen/CCSNet_publish_dataset/{name}.hdf5', 'r')
    variable = np.array(hf_r.get(name))
    hf_r.close()
    return variable
```

```python
test_x = get_dataset('test_x')
test_y_BPR = get_dataset('test_y_BPR')
test_y_P_init = get_dataset('test_y_P_init')
test_y_P = test_y_BPR - test_y_P_init
test_y_SG = get_dataset('test_y_SG')
test_y_BXMF = get_dataset('test_y_BXMF')
test_y_BYMF = get_dataset('test_y_BYMF')
test_y_BDENG = get_dataset('test_y_BDENG')
test_y_BDENW = get_dataset('test_y_BDENW')
```

```python
SG_model = tf.keras.models.load_model('trained_models/SG_v1.h5', 
                                      custom_objects={'ReflectionPadding3D':ReflectionPadding3D}, compile=False)
```

```python
dP_model = tf.keras.models.load_model('trained_models/dP_v1.h5', 
                                      custom_objects={'ReflectionPadding3D':ReflectionPadding3D}, compile=False)
```

```python
BXMF_model = tf.keras.models.load_model('trained_models/bxmf_v1.h5', 
                                      custom_objects={'ReflectionPadding3D':ReflectionPadding3D}, compile=False)
```

```python
BYMF_model = tf.keras.models.load_model('trained_models/bymf_v1.h5', 
                                      custom_objects={'ReflectionPadding3D':ReflectionPadding3D}, compile=False)
```

```python
BDENG_model = tf.keras.models.load_model('trained_models/bdeng_v1.h5', 
                                      custom_objects={'ReflectionPadding3D':ReflectionPadding3D}, compile=False)
```

```python
BDENW_model = tf.keras.models.load_model('trained_models/bdenw_v1.h5', 
                                      custom_objects={'ReflectionPadding3D':ReflectionPadding3D}, compile=False)
```

```python
dnorm_temp = lambda a : (a ) * (180 - 30) + 30
dnorm_P = lambda a : (a) * (300 - 100) + 100

def make_SG(x):
    sg = SG_model.predict(x)
    return sg[0,:,:,:,0]

def make_dP(x):
    dp = dP_model.predict(x) * 300
    return dp[0,:,:,:,0]

def make_BXMF(x, dp, sg):
    data_x_and_pred = np.concatenate([x, 
                                      sg[np.newaxis,...,np.newaxis], 
                                      dp[np.newaxis,...,np.newaxis]/600], axis=-1)
    bxmf = BXMF_model.predict(data_x_and_pred) * 0.038
    return bxmf[0,:,:,:,0]

def make_BYMF(x, bpr, sg):
    temp = np.repeat(x[:,:,:,-4,:][:,:,:,np.newaxis,:], 24, axis=-2)
    x_input = np.concatenate([temp, (bpr - 100)/(565 - 100), sg], axis=-1)
    y = BYMF_model.predict(x_input)*0.1+0.9
    y[sg == 0] = 0
    return y

def make_BDENG(x, bpr, bymf):
    temp = np.repeat(x[:,:,:,-4,:][:,:,:,np.newaxis,:], 24, axis=-2)
    bymf_x = np.copy(bymf)
    bymf_x[bymf_x == 0] = 0.9
    x_input = np.concatenate([temp, (bpr - 100)/(565 - 100), (bymf_x-0.9)/0.1], axis=-1)
    y = BDENG_model.predict(x_input) * 900 + 100
    y[bymf == 0] = 0
    return y

def make_BDENW(x, bpr, bxmf):
    temp = np.repeat(x[:,:,:,-4,:][:,:,:,np.newaxis,:], 24, axis=-2)
    x_input = np.concatenate([temp, (bpr - 100)/(565 - 100), bxmf/0.038], axis=-1)
    y = BDENW_model.predict(x_input) * 400 + 700
    y[bxmf == 0] = 0
    return y

def make_BPR(x, dp):
    temperature = dnorm_temp(x[0, 0, 0, -4, 0])
    pressure =  dnorm_P(x[0, 0, 0, -3, 0])
    pinit = get_p_init(temperature, pressure)
    return dp + pinit[...,0]
```

```python
def make_BXMF(x, p, sg):
    data_x_and_pred = np.concatenate([x, 
                                      sg[np.newaxis,...,np.newaxis], 
                                      p[np.newaxis,...,np.newaxis]/600], axis=-1)
    bxmf = BXMF_model.predict(data_x_and_pred) * 0.038
    return bxmf[0,:,:,:,0]
```

```python
def get_p_init(t, p):
    print(t,p)
    tt = np.array((float(t) - 30)/(170 - 30)).reshape((1, 1))
    pp = np.array((float(p) - 100) / (370 - 100)).reshape((1, 1))
    rho_w = PropsSI("D", "T", tt + 273.15, "P", pp * 1e5, "water") 
    print(t,p)
    print(rho_w)
    p_same_rho = []
    for i in range(96):
        p_bottom = p + rho_w * 2.083330 * 9.8 / 100000
        p_same_rho.append((p_bottom + p) / 2)
        p = p_bottom
    p_init = np.array(p_same_rho)
    p_init = np.repeat(p_init[:, np.newaxis], 200, axis=1)
    return p_init
```

```python
def time_CCSNet(x):
    sg = make_SG(x)
    dp = make_dP(x)
    bpr = make_BPR(x, dp)
    bxmf = make_BXMF(x, bpr, sg)
    bymf = make_BYMF(x, bpr[np.newaxis,:,:,:,np.newaxis], sg[np.newaxis,:,:,:,np.newaxis])[0,...,0]
    bdenw = make_BDENW(x, bpr[np.newaxis,:,:,:,np.newaxis], bxmf[np.newaxis,:,:,:,np.newaxis])[0,...,0]
    bdeng = make_BDENG(x, bpr[np.newaxis,:,:,:,np.newaxis], bymf[np.newaxis,:,:,:,np.newaxis])[0,...,0]
    return sg, dp, bxmf, bymf, bdenw, bdeng
```
